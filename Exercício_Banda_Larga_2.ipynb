```python
import pandas as pd
```

```python
df = pd.read_csv("https://storage.googleapis.com/dados_turmabc6_soul/Dados/Acessos_Banda_Larga_Fixa_2021.csv", sep=';')
df.head()
```

```python
pd.unique(df['Meio de Acesso'])
```

```python
pd.unique(df['Tecnologia'])
```

```python
#Localizando e modificando - Onde tiver OUTROS mudar para CONSORCIO
df.loc[df['Grupo Econômico'] == "OUTROS", ['Grupo Econômico']] = 'CONSORCIO'
```

```python
#Localizando e modificando - Onde tiver Indeterminado mudar para AR
df.loc[df['Meio de Acesso'] == "Indeterminado", ['Meio de Acesso']] = 'AR'
```

```python
#Localizando e modificando - Onde tiver "LTE|GSM|GPRS|OFDMA/TDD|EDGE" mudar para '2G/3G/4G'

df.loc[df.Tecnologia == "LTE", ['Tecnologia']] = '2G/3G/4G'
df.loc[df.Tecnologia == "GSM", ['Tecnologia']] = '2G/3G/4G'
df.loc[df.Tecnologia == "GPRS", ['Tecnologia']] = '2G/3G/4G'
df.loc[df.Tecnologia == "OFDMA/TDD", ['Tecnologia']] = '2G/3G/4G'
df.loc[df.Tecnologia == "EDGE", ['Tecnologia']] = '2G/3G/4G'

#Localizando e modificando - Onde tiver "FTTH", "FTTB" mudar para FIBRA
df.loc[df.Tecnologia == "FTTH", ['Tecnologia']] = 'FIBRA'
df.loc[df.Tecnologia == "FTTH", ['Tecnologia']] = 'FIBRA'
```

```python
#Filtrando para saber se a palavra está contida no dataframe
filtro = df.Tecnologia.str.contains('FIBRA')
df.loc[filtro]
```

```python
#Filtrando duas colunas para poder interligar
filter1 = df.UF.isin(['PB', 'CE', 'RN'])
filter2 = df['Porte da Prestadora'] != 'Pequeno Porte'
```

```python
#Interligando duas colunas em apenas uma consulta
df.loc[filter1 & filter2]
```
