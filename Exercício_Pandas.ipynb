No dataframe padrão , criar 3 colunas de "backup" sendo 1 para UF , outra para cidade e outra para aerodromo

"normalize" os **** do aerodromo e troque para MODF , 
Onde existir a cidade de Rio de Janeiro troque a classificação do acidente para TRAGEDIA e onde existir o aerodromo SBPK troque a cidade para ESPECIAL

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/dados/ocorrencia_2010_2020 - ocorrencia_2010_2020.csv", sep=";", parse_dates = ["ocorrencia_dia"], dayfirst = True) #parse_dates transforma uma coluna em tipo data
df
```

```python
df.drop(['codigo_ocorrencia2'], axis = 1, inplace = True)
df
```

```python
df['backup_uf'] = df.ocorrencia_uf
```

```python
df['backup_cidade'] = df.ocorrencia_cidade
```

```python
df['backup_aerodromo'] = df.ocorrencia_aerodromo
```

```python
df.head()
```

```python
#Localiza e modifica
df.loc[df.ocorrencia_aerodromo == "****", ['ocorrencia_aerodromo']] = 'MODF'
```

```python
df.loc[df.ocorrencia_cidade == "RIO DE JANEIRO", ['ocorrencia_classificacao']] = 'TRAGEDIA'
```

```python
df.loc[df.ocorrencia_aerodromo == "SBPK", ['ocorrencia_cidade']] = 'ESPECIAL'
```
