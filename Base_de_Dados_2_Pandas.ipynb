```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/dados/cadastro.csv", parse_dates = ["earliest priority date", "application date", "publication date", "publication year"], dayfirst = True)
df.head()
```

```python
df.dtypes
```

```python
df.rename(columns={'row no': 'id',
                   'publication number': 'numero_da_publicacao',
                   'kind code': 'tipo_do_codigo',
                   'title': 'titulo',
                   'priority document': 'documento_prioritario',
                   'priority country': 'pais_prioritario',
                   'earliest priority date': 'primeira_data_de_prioridade',
                   'earliest priority year': 'primeiro_ano_de_prioridade',
                   'application date': 'data_de_aplicacao',
                   'publication date': 'data_de_publicacao',
                   'publication year': 'ano_de_publicacao',
                   'applicant': 'candidato',
                   'inventors': 'inventores',
                   'ipc classes': 'aulas_ipc',
                   'family members': 'membros_da_familia',
                   'family size': 'tamanho_familia',
                   'wo in family': 'wo_na_familia',
                   'grant or application type families': 'concessao_ou_tipo_de_aplicacao_das_familias',
                   'granted patent': 'patente_concedida'
                   }, inplace = True)
```

```python
df.head()
```

```python
#Transformando a coluna id em rótulo
df.set_index('id', inplace = True) #inplace = True valida o que será setado
```

```python
df.head()
```

```python
df.drop(['numero_da_publicacao', 'primeira_data_de_prioridade', 'aulas_ipc', 'membros_da_familia', 'wo_na_familia'], axis = 1, inplace = True)
```
