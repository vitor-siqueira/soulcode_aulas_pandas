Deve ser convertido neste dataframe os numeros em strings que façam sentido Ex: sexo , raça, entre outros. As colunas que não façam sentido podem ser dropadas

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/dados/pesquisa.csv")
df
```

```python
#Organizando o id em números crescentes
df = df.sort_values(by=['id'])
df.head()
```

```python
#Transformando a coluna id em rótulo
df.set_index('id', inplace = True) #inplace = True valida o que será setado
```

```python
df[['sexo', 'raca', 'origem']] = df[['sexo', 'raca', 'origem']].astype(str)
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Localiza e modifica

#sexo
df.loc[df.sexo == "1", ['sexo']] = 'masculino'
df.loc[df.sexo == "0", ['sexo']] = 'feminino'

#raça
df.loc[df.raca == "1", ['raca']] = 'indigina/amarelo'
df.loc[df.raca == "2", ['raca']] = 'negro'
df.loc[df.raca == "3", ['raca']] = 'branco'
df.loc[df.raca == "4", ['raca']] = 'pardo'

#origem
df.loc[df.origem == "1", ['origem']] = 'norte/nordeste'
df.loc[df.origem == "2", ['origem']] = 'centro-oeste/sudeste'
df.loc[df.origem == "3", ['origem']] = 'sul'
```

```python
df.head()
```

```python
df.drop(['tipo', 'prog'], axis = 1, inplace = True)
```
